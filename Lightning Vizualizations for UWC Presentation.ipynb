```python
pointlist = pd.read_csv('/Users/millerc/Dropbox/03-ETH/30-ResearchProposal/Presentation/ForceLayoutFiles/UWC Master List for JSON Conversion - reducedVAV.csv')
catlist1 = list(pointlist.Category.unique())
taglist2 = list(pointlist.PointTag.unique())
```

```python
pointlist
```

```python
adjdict = {}
for category in catlist1:
    tagdict = {}
    for tag in pointlist[(pointlist.Category == category)].PointTag.unique():
        tagdict[tag] = pointlist[(pointlist.Category == "Meters") & (pointlist.PointTag == "WaterMeter")].EquipTag.to_dict()
    adjdict[category] = tagdict
```

```python
adjdict
```

```python
G = nx.from_dict_of_dicts(adjdict)
mat = nx.adjacency_matrix(G).todense()
lgn.force(mat)
```
